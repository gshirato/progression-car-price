## Developing a Model

```python
def preprocessing(df: pd.DataFrame) -> pd.DataFrame:
    res = df.copy()
    res['clean_title'] = res['clean_title'].fillna('NA')
    res['is_clean_title'] = res['clean_title'] == 'Yes'
    res['age'] = (2024 - res['model_year']).map(lambda x: max(x, 1))
    res['milage_per_year'] = res['milage'] / res['age']
    res['had_accident'] = res['accident'] == 'At least 1 accident or damage reported'
    return res
```

```python
from sklearn.model_selection import train_test_split
preprocessed = preprocessing(df)
train, val = train_test_split(preprocessed, test_size=0.3, random_state=42)
```

```python
def split_X_y(df, X_names, y_name):
    return df[X_names], df[y_name]
```

```python
X_train, y_train = split_X_y(train, ['is_clean_title', 'milage_per_year', 'had_accident'], 'price')
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_estimators=100, random_state=42)
forest.fit(X_train, y_train)
```

```python
X_val, y_val = split_X_y(val, ['is_clean_title', 'milage_per_year', 'had_accident'], 'price')
```

```python
y_pred = forest.predict(X_val)
```

```python
result = pd.DataFrame({"y_pred": y_pred, 'y_val': y_val})
result['diff'] = result['y_pred'] - result['y_val']
```

```python
result
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_val, y_pred))
```
